with seaborn

```python
import seaborn as sns
```

```python
sns.countplot(data=listings_df, x="neighbourhood_group")
```

```python
sns.countplot(data=listings_df, x="room_type")
```

## What's the influene of the neighbourhood group on room type

```python
listings_df["room_type"].unique()
```

```python
sns.countplot(data=listings_df, x="neighbourhood_group", order=order, hue="room_type")
```

## What is the distribution of flat prices

```python
sns.displot(listings_df["price"])
```

Not very readable

```python
listings_df[listings_df["price"] > 500].shape
```

```python
affordable_df = listings_df[listings_df["price"] <= 500]
```

```python
sns.displot(affordable_df["price"])
```

Why spikes? probably human psychology, 199 seems a lot cheaper that 200

## What is the distribution of flat prices based on the neighbourhood group?

```python
sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15, 8))
sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

## Can we plot the listings on the map?

we'll use a scatterplot: y=latitude, x=longitude

```python
affordable_df.head()
```

```python
affordable_df.plot(
    kind="scatter",
    x="longitude",
    y="latitude",
    c="price",
    cmap="inferno",
    colorbar=True,
    alpha=0.8,
    figsize=(12, 8)
)
```
